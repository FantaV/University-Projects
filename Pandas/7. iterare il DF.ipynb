```python
# 7. ITERARE I DATAFRAME

# per iterare si utilizzano i metodi ITEMS, ITERROWS, ITERTUPLES

import pandas as pd

df = pd.read_csv("tsla.csv").head(10)
```

```python
# ITEMS: si inserisce la KEY e il VALUE. Va a prendere la riga come se fosse un DICT 

for key,value in df.items():
  print(key,value)  # vengono mandati a schermo i primi dieci indici con i calori di ogni colonna dedicata 
```

```python
# ITERROWS: dobbiamo inserire l'INDEX e la ROW

for index,row in df.iterrows():
    print(index,row)
```

```python
# ITERTUPLES si usa solo ROW

for row in df.itertuples():
    print(row)
```

```python
# N.B.: non si devono modificare i dati quando iteriamo perché l'iterazione ci va a creare delle copie anziché delle view. Iterare serve per leggere i file 

# esempio di MODIFICA

df2 = pd.read_csv("tsla.csv").head(1)

for index,row in df2.iterrows():
    if(index==0):
        row['Open']= ' Giocolandia'
    print(row)

print(df2.head(1))

# pur avendo gli stessi risultati la colonna open riporta due nomi diversi, e quello modificato viene preso correttamente. Questo perché nel FOR viene effettuata una copia staccata.
# Quindi il nuovo array e proprietario degli oggetti all'interno di essi 

# Se invece fosse stata una view avrebbe modificato anche l'originale
```
