```python
# RAGGRUPPARE CON GROUPBY

import pandas as pd 

df = pd.read_csv('pokemon.csv')
```

```python
# RAGGRUPPARE PER COLONNA

types = df.groupby('Type 1')

print(types) # senza specificare ci andr√† a creare solo il gruppo mentre con:

print(types.groups)
```

```python
# considerato il caos di sopra dobbiamo CICLARE i nostri elementi 

for name,group in types:
    print(name)
    print(group)
```

```python
# RAGGRUPPARE MULTIPLE COLONNE 

typesM = df.groupby(['Type 1', 'Type 2'])

for name,group in typesM:
    print(name)
```

```python
# ITERARE I GRUPPI

for name,group in typesM:
    print(name)
    print(group)
```

```python
# GENERARE DELLE STATISTICHE PI√ô AVANZATE: sum, mean, count, min, max, e aggiungere colonna count

# esempio con MEAN + COLONNA COUNT (in modo da decidere quali colonne mandare a schermo)

types = df.groupby(['Type 1']).mean(numeric_only=True)[['HP','Attack','Defense','Speed']].sort_values(by='Attack', ascending=False)        # numeric_only = True OBBLIGATORIO 

print(types)
```

```python
# MIN e MAX

types = df.groupby(['Type 1']).min(numeric_only=True)[['HP','Attack','Defense','Speed']].sort_values(by='HP', ascending=False)

print(types)

types = df.groupby(['Type 1']).max(numeric_only=True)[['HP','Attack','Defense','Speed']].sort_values(by='Attack', ascending=False)

print(types)
```

```python
# COUNT

types = df.groupby(['Type 1']).count()

print(types)
```

```python
# SCEGLIERE LA COLONNA CON COUNT: bisogna prima creare una variabile count 

df['count'] = 1

types = df.groupby(['Type 1']).count()['count']

print(types)
```

```python
# magari voglio vedere anche le combinazioni

types = df.groupby(['Type 1','Type 2']).count()['count']

print(types)
```
