```python
# 3. BAR CHARTS 

from matplotlib import pyplot as plt
import numpy as np 
```

```python
x = [25,26,29,31,32,33,36]
y = [6,9,8,10,11,9,10]
y2 = [9,10,9,8,10,12,11]  
```

```python
# GRAFICO VERTICALE: .BAR

plt.bar(x,y, label='Italiani', color='red') 
plt.bar(x,y2, label='Africani', color='#212121')

plt.show()
```

```python
# BARRE AFFIANCATE: per avere barre affiancate bisogna usare NUMPY come escamotage dato matplot non ha una vera e proprio funzione a riguardo 

# innanzitutto si devono creare degli indici

indexes = np.arange(7)

plt.bar(indexes,y, label='Italiani', color='red')     # porto gli indici al posto della x
plt.bar(indexes,y2, label='Africani', color='#212121')

# però bisogna rendere indixes pari ai nostri valori e bisogna introdurre il concetto di LARGHEZZA  con la variabile WIDHT 

width = 0.3

plt.bar(indexes,y, label='Italiani',width=width, color='red')
plt.bar(indexes,y2, label='Africani',width=width, color='#212121')

plt.show()
```

```python
# ed inserendo indixes + width:

plt.bar(indexes+width,y, label='Italiani',width=width, color='red')
plt.bar(indexes,y2, label='Africani',width=width, color='#212121')

plt.show()
```

```python
# per aggiustare i TICKS e centrarli ed aggiungere i nostri valori x

plt.xticks(indexes+width/2, x)      # tick dell'asse x

plt.show()
```

```python
# GRAFICO ORIZZONTALE: .barh

plt.barh(indexes,y, label='Italiani', color='red')  
plt.barh(indexes,y2, label='Africani', color='#212121')

plt.show()
```
