```python
# 4. GRAFICO A TORTA (PIE CHARTS)

from matplotlib import pyplot as plt 
```

```python
# CREARE GRAFICO BASE CON .pie

slices = [25,26,19,31,33,12,27]



labels = ['Abruzzo', 'Marche', 'Basilicata','Campania','Toscana','Molise','Puglia']     # definire le etichette
colors = ['red', 'brown','green','blue','orange','yellow','magenta']                    # definire i colori
explode = [0,0,0.12,0,0,0,0.1]                                                          # separare uno slice di torta 

plt.figure(figsize=(5, 5))  # Cambia le dimensioni della figura

plt.pie(slices, 
        labels=labels, 
        colors=colors, 
        wedgeprops={'edgecolor': '#000000'},  # EDGECOLOR per i bordi
        startangle=180,                       # STARTANGLE per l'angolo
        explode=explode,         
        autopct='%1.1f%%')                    # AUTOPCT per l'utilizzo delle percentuali ai valori, e la sintassi richiede la FORMATSTRING 
 



plt.style.use('ggplot')
plt.title('Nab. in Mil')
plt.show()
```
