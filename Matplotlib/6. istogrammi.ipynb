```python
# 6. ISTOGRAMMI 

from matplotlib import pyplot as plt
```

```python
# BINS: sono delle scatole in cui aggreghiamo i dati che abbiamo

x = [25,25,26,27,27,29,31,32,33,36,37,40,41,41,41,43,55,56,59,61,63,64,64,67,68,68]

plt.hist(x, bins=8) # con bins=5 mi ha creato un grafico hist calcolando quante unità ci sono per ogni categoria

plt.plot()
```

```python
# ASSOCIARE BINS SPECIFICI
# EDGECOLOR per i bordi 
# AXVLINE per mostrare una linea verticla nel grafico utilizzandola ad esempio come media (ax= axis, vline= vertical line)

x = [25,25,26,27,27,29,31,32,33,36,37,40,41,41,41,43,55,56,59,61,63,64,64,67,68,68]
sumX = sum(x)
lenghtX = len(x)
meanX = sumX/lenghtX

bins = [10,20,30,40,50,60,70] # creiamo noi delle fasce d'età

plt.hist(x,bins=bins, color='orange', edgecolor='black')       # edgecolor per i bordi 
plt.axvline(meanX, color='red', linewidth=2)

plt.plot()
```
