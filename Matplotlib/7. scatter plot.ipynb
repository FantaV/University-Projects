```python
# 7. SCATTER PLOT (Grafico a dispersione)

from matplotlib import pyplot as plt
```

```python
# SCATTER pr utilizzare il grafico a dispersione
# S: per modificare la size dei punti
# C: color



x = [25,25,26,27,27,29,31,32,33,36,37,40,41,41,41,43,55,56,59,61,63,64,64,67,68,68]
y = [5,9,7,10,11,9,12,14,8,3,8,9,10,7,12,11,13,10,12,15,12,11,9,8,10,13]


plt.scatter(x,y, 
            s=100,     # size dei punti
            color='green', 
            marker='X',     # per scegliere come visualizzare il dato
            edgecolor='black',
            linewidth=2,
            alpha=0.5       # alpha vale sia per i bordi del punto che il riempimento interno
            )

plt.grid(linewidth=0.3)
plt.style.use('fast')

plt.title('Et√†')
plt.show()
```

```python
# CON IL GRAFICO A DISPERSIONE POSSIAMO AVERE ANCHE DEI PUNTI PI√ô SCURI ED ALTRI PI√ô CHIARI che va ad indicare dei dati specifici

x = [25,25,26,27,27,29,31,32,33,36,37,40,41,47,46,43,55,56,59,61,63,64,64,67,68,68]
y = [5,9,7,10,11,9,12,14,8,3,8,9,10,7,12,11,13,10,12,15,12,11,9,8,10,13]

colors = [x]
sizes = [100,110,120,130,140,150,160,170,180,190,200,210,220,230,240,250,260,270,280,290,300,310,320,330,340,350] # per cambiare le size di ogni punto se vogliamo vederli 

plt.scatter(x,y,
            s=sizes, 
            edgecolor='black',
            linewidth=1,      
            cmap='Accent',    # color map che ci restituisce la scala colori che vogliamo 
            c=colors)

cbar = plt.colorbar()         # 
cbar.set_label('Intensity')   # label che indica l'intensit√† del colore in base ai miei dati



plt.grid(linewidth=0.3)
plt.style.use('fast')

plt.title('Et√†')
plt.show()
```
