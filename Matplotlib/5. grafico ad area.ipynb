```python
# 5. GRAFICO A LINEE RIEMPITO

from matplotlib import pyplot as plt
import numpy as np
```

```python
# RIEMPIRE L'AREA

x = [25,26,29,31,32,33,36]
y = [5,9,7,10,11,9,12]

plt.plot(x,y)
plt.fill_between(x,y, color= 'orange', alpha=0.5)      # riempire l'area usando FILL_BETWEEN e ALPHA per l'opacità
```

```python
# RIEMPIRE IN PUNTI SPECIFICI 

media = 10

plt.fill_between(x,y,media, color='blue', alpha=0.5)

plt.plot()
```

```python
# AGGIUNGERE CONDIZIONI 

x = np.array([25,26,29,31,32,33,36])
y = np.array([5,9,7,10,11,9,12])
media = 10

plt.fill_between(x,y,media, where=(y>9), color='blue', alpha=0.5, interpolate=True)

plt.plot()
```
