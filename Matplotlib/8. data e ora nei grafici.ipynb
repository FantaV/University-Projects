```python
# 8. DATA E ORA NEI GRAFICI 

from datetime import datetime, timedelta
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates
```

```python
x = [
    datetime(2022,1,10),
    datetime(2022,1,15),
    datetime(2022,1,20),
    datetime(2022,1,25),
    datetime(2022,1,30)
]

y = [3,6,2,8,5]

plt.plot_date(x,y,
              linestyle='solid')
# rendere le date più leggibili 
plt.gcf().autofmt_xdate() # get current figure (prendi la figura attuale e fammi questo. La funzione è ciò che inseriamo dopo la parentesi .autofmt_xdate)

# sistemare le date
date_formattate = mpl_dates.DateFormatter('%d %b %Y')
plt.gca().xaxis.set_major_formatter(date_formattate)  # get current axis e fammi questo, mi andrà a stirare le date inserendo JAN 

plt.grid(linewidth=0.4)
plt.title('Date a 5')
plt.show()
```
