```python
# 2. PERSONALIZZARE GRAFICI

from matplotlib import pyplot as plt
```

```python
x = [25,26,29,31,32,33,36]
y = [6,9,8,10,11,9,10]
```

```python
# AGGIUNGERE TITOLO, ETICHETTE, SECONDA LINEA, LEGENDA

# AGGIUNGERE GRIGLIA, COLORE, DIMENSIONE, STILE LINEA, MARCATORI E STILE GRAFICO

plt.title('Wonderful')  # titolo
plt.xlabel('Età')       # xlabel
plt.ylabel('Abitanti in Mil')  # ylabel 

y2 = [9,10,9,8,10,12,11]    # seconda linea

# LEGEND: è sempre più funzionale inserire la leggenda all'interno della linea di riferimento poiché se ne avessimo molte non riusciremmo a riconoscerle

plt.plot(x,y, label='Italiani', color='red', linewidth=3, linestyle='--', marker='o')  # STILE LINEA E MARCATORI 

plt.plot(x,y2, label='Africani', color='#212121', linestyle='dotted', marker='.')   # i colori possono essere scritti per estesi o in formato decimale

plt.legend()        # importante plottare anche legend per mostrarle correttamente
plt.grid()          # griglia

# STILE GRAFICO: per avere la lista completa possiamo: 
print(plt.style.available)
plt.style.use('seaborn-v0_8')  # per usarli

plt.show()
```
