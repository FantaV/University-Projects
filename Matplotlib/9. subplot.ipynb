```python
# 9. SUBPLOT 

# i subplot ci danno la possibilità di avere più grafici su un'unica finestra oppure di creare più finestre

from matplotlib import pyplot as plt
```

```python
x = [25,25,26,27,27,29,31,32,33,36,37,40,41,42,43,44,55,56,59,61,63,64,64,67,68,68]
y = [5,9,7,10,11,9,12,14,8,3,8,9,10,7,12,11,13,10,12,15,12,11,9,8,10,13]

# FIGURE e AXIS per lavorare con più figure

fig, ax = plt.subplots()        # FIG sta per la finestra, mentre AX sta per il grafico

ax.plot(x,y)

# la sintassi del subplot cerca il set_
ax.set_title('Grafico')
ax.set_xlabel("Fasce d'età")

plt.show() # non è più lo stesso plt ma abbiamo creato un oggetto a parte che stiamo utilizzando per il grafico mentre gli AX fanno riferimento ad AX 
```

```python
# CREARE PIÙ GRAFICI IN UNA FINESTRA

# UNPACKING (spacchettare): possiamo spacchetare gli AX e renderli relativamente uguali ai loro assi di riferimento
# ricordare: gli ax sono i nostri grafici e con ax1 e ax2 vuol dire che sono i nostri due grafici

fig, (gr1, gr2) = plt.subplots(nrows=2, ncols=1,        # (ax1, ax2) = [<Axes: > <Axes: >]
                               sharex=True)             # ci permette di non ripetere il numero sull'asse ma di inserirlo solo per il secondo grafico 

x = [25,25,26,27,27,29,31,32,33,36,37,40,41,42,43,44,55,56,59,61,63,64,64,67,68,68]
y = [5,9,7,10,11,9,12,14,8,3,8,9,10,7,12,11,13,10,12,15,12,11,9,8,10,13]

gr1.plot(x,y)
gr2.plot(x,y)

gr1.set_title('Grafico')

gr2.set_xlabel("Fasce d'età")       # in teoria possiamo anche non inserire per entrambi i grafici le rispettive label ed title 

plt.style.use('classic')
```

```python
# CREARE PIÙ FINESTRE

fig1, (gr1a, gr2b) = plt.subplots(nrows=2, ncols=1, sharex=True) # le lavorazioni sul grafico sono indipendenti dall'altro presente nell'altra finestra 
fig2, gr2b = plt.subplots() 

gr1a.set_title('Grafico')
gr1a.set_xlabel("Fasce d'età")
gr1a.plot(x,y)

gr2b.set_title('Grafico')
gr2b.set_xlabel("Fasce d'età")
gr2b.plot(x,y)

plt.show()
```
